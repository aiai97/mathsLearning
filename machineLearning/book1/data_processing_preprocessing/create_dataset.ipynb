create datasets  make()
make_blobs: Generates a multi-class clustering dataset where each class is represented by a cluster of points generated from Gaussian distributions.

make_classification: Generates a classification dataset with specified classes and features. The dataset's feature count, class count, and sample count can be controlled by adjusting the parameters.

make_circles: Generates a binary classification dataset with a circular decision boundary between the two classes.

make_moons: Generates a binary classification dataset with a non-linear decision boundary resembling two crescent moons.

make_regression: Generates a regression dataset with a linear relationship. The sample count, feature count, and noise level can be controlled by adjusting the parameters

```python
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
from sklearn.datasets._samples_generator import make_blobs
import numpy as np
```

```python
center = [[1,1],[-1,-1],[1,-1]]
# centers : int or ndarray of shape (n_centers, n_features), default=None
# The number of centers to generate, or the fixed center locations.
# If n_samples is an int and centers is None, 3 centers are generated.
# If n_samples is array-like, centers must be
# either None or an array of length equal to the length of n_samples.
X,labels = make_blobs(n_samples=200,centers=center,n_features=2,cluster_std=0.2)
df = pd.DataFrame(np.c_[X,labels],columns=['feature1','feature2','labels'])
df.plot.scatter('feature1','feature2',s=100,c=list(df['labels']),cmap='rainbow',
                colorbar=False,alpha=0.8,title='dataset by make_blobs')
plt.show()
```
