Bermoulli distribution

```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 10
p = 0.5
#generate a sequence of values with a specified start, stop, and step size. It returns a one-dimensional array
k = np.arange(0,10)
```

P(X = k) = p^k(1 - p)^(1 - k)

binom refers to the binomial distribution class or object from a library, such as SciPy.
pmf is a method or function provided by the binomial distribution class. It stands for "probability mass function" and is used to calculate the probability of obtaining a specific value or set of values in a binomial distribution.
k represents the number of successful outcomes or "successes" in the binomial distribution. It can be a single value or an array of values.
n is the total number of trials or observations in the binomial distribution.
p is the probability of success for each trial in the binomial distribution.


```python
binomial = binom.pmf(k,n,p)
plt.plot(k,binomial)
#%i is a placeholder for an integer value, which will be replaced by the value of n.
#%0.2f is a placeholder for a floating-point value with two decimal places, which will be replaced by the value of p.
#(n, p) specifies the variables n and p to be inserted into the title string.
plt.title('Binomial: n = %i,p =%0.2f'%(n,p),fontsize=15)
plt.xlabel('Number of successes')
plt.ylabel('Probability of successes',fontsize=15)
plt.show()
```

Poisson distribution
P(X = k) = (e^(-λ) * λ^k) / k!
- `P(X = k)` represents the probability that the random variable `X` takes on the value `k`.
- `e` is the mathematical constant approximately equal to 2.71828.
- `λ` is the average rate or expected number of events occurring in the interval.
- `k` is the number of events occurring in the interval.
- `k!` denotes the factorial of `k`, which is the product of all positive integers less than or equal to `k`.


Let's say you work at a call center and receive an average of 5 customer calls per minute. You want to determine the probability of receiving exactly 3 calls in a given minute.

To solve this problem using the Poisson distribution, you can use the formula:

P(X = k) = (e^(-λ) * λ^k) / k!
where λ is the average rate of events (in this case, the average number of customer calls per minute) and k is the number of events you are interested in (in this case, 3 calls).

Plugging in the values, we have:

λ = 5 (average calls per minute)
k = 3 (desired number of calls)
Using the formula, we can calculate the probability as follows:

P(X = 3) = (e^(-5) * 5^3) / 3!
         ≈ 0.140
Therefore, the probability of receiving exactly 3 customer calls in a minute is approximately 0.140, or 14.0%.

poisson(lam=1.0, size=None)
    Draw samples from a Poisson distribution.
lam : float or array_like of floats
Expected number of events occurring in a fixed-time interval,
must be >= 0. A sequence must be broadcastable over the requested
size.
size : int or tuple of ints, optional
Output shape.  If the given shape is, e.g., ``(m, n, k)``, then
``m * n * k`` samples are drawn.  If size is ``None`` (default),
a single value is returned if ``lam`` is a scalar. Otherwise,
``np.array(lam).size`` samples are drawn.

```python
X = np.random.poisson(lam=5,size=10000)
# X is the data or values that will be used to construct the histogram.
# bins=15 specifies the number of bins or bars to be used in the histogram. In this case, there will be 15 bins.
# range=[0,15] sets the range of values to be considered for the histogram. Only values between 0 and 15 (inclusive) will be plotted.
# color='g' sets the color of the histogram bars to green ('g' represents green).
# alpha=0.5 sets the transparency level of the histogram bars. In this case, the bars will be semi-transparent, with an alpha value of 0.5.
s = plt.hist(X,bins=15,range=[0,15],color='g',alpha=0.5)
# s[1][0:15] represents the x-values or the bin edges from the histogram plot. s[1] refers to the bin edges stored in the variable s (returned by the plt.hist() function), and [0:15] specifies that only the first 15 elements of the bin edges will be used.
# s[0] represents the y-values or the heights of the histogram bins. s[0] refers to the heights stored in the variable s.
plt.plot(s[1][0:15],s[0],'r')
plt.grid()
plt.show()
```
